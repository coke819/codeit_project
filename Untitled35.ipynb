```python
import numpy as np

# Step 1. did_recharge 값 클리닝
DF_eligible['did_recharge_clean'] = DF_eligible['did_recharge'].map({'0': False, 'True': True, 'False': False})

# Step 2. 이상치 탐색을 위한 IQR 기준 계산 함수
def detect_outlier_iqr(series):
    q1, q3 = np.percentile(series, [25, 75])
    iqr = q3 - q1
    upper_bound = q3 + 1.5 * iqr
    return upper_bound

# event_count 이상치 경계 계산
event_upper_bound = detect_outlier_iqr(DF_eligible['event_count'])

# friend_count 이상치 경계 계산
friend_upper_bound = detect_outlier_iqr(DF_eligible['friend_count'])

# 이상치 flag 컬럼 생성
DF_eligible['event_outlier'] = DF_eligible['event_count'] > event_upper_bound
DF_eligible['friend_outlier'] = DF_eligible['friend_count'] > friend_upper_bound

# 이상치 개수 확인
event_outliers = DF_eligible['event_outlier'].sum()
friend_outliers = DF_eligible['friend_outlier'].sum()

print("event_count 이상치 기준:", event_upper_bound, f"({event_outliers}건)")
print("friend_count 이상치 기준:", friend_upper_bound, f"({friend_outliers}건)")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

def extract_aha_moment_robust(DF, behavior_pivot,
                               df_behavior_features=['event_count', 'friend_count', 'did_recharge_clean'],
                               min_occurrence=0.01, gap_threshold=1e-4, top_k=5):

    # 1. 데이터 병합
    merged_df = pd.merge(DF, behavior_pivot, on='user_id', how='inner')

    # 2. feature 리스트 구성
    behavior_features = list(behavior_pivot.columns.drop('user_id'))
    behavior_cols = df_behavior_features + behavior_features

    aha_summary_list = []
    cluster_list = merged_df['cluster'].unique()

    for c in cluster_list:
        cluster_df = merged_df[merged_df['cluster'] == c]

        # 희소 feature 제거
        feature_activity = (cluster_df[behavior_cols] > 0).mean()
        valid_features = feature_activity[feature_activity >= min_occurrence].index.tolist()
        if len(valid_features) == 0:
            continue

        # group별 분리
        retained_df = cluster_df[cluster_df['is_retained'] == 1]
        churned_df = cluster_df[cluster_df['is_retained'] == 0]

        records = []
        for feature in valid_features:
            # 평균 계산
            churned_mean = churned_df[feature].mean()
            retained_mean = retained_df[feature].mean()
            gap = retained_mean - churned_mean

            if abs(gap) < gap_threshold:
                continue

            # 통계검정 (t-test)
            try:
                stat, p_value = ttest_ind(retained_df[feature], churned_df[feature], equal_var=False, nan_policy='omit')
            except:
                p_value = 1.0

            # 종합 점수: gap / p_value → 유의한 큰 gap에 가중치
            score = abs(gap) / (p_value + 1e-8)

            records.append({
                'cluster': c,
                'feature': feature,
                'Churned': churned_mean,
                'Retained': retained_mean,
                'gap': gap,
                'p_value': p_value,
                'score': score
            })

        if records:
            cluster_result = pd.DataFrame(records).sort_values('score', ascending=False).head(top_k)
            aha_summary_list.append(cluster_result)

    # 통합 결과
    aha_summary_df = pd.concat(aha_summary_list, ignore_index=True)
    return aha_summary_df
```

```python
aha_result = extract_aha_moment_robust(DF, behavior_pivot)

# 결과 확인
aha_result
```

```python
aha_result_filtered = aha_result[aha_result['p_value'] <= 0.05].sort_values('score', ascending=False)
aha_result_filtered
```

```python
# epsilon 방어용 (0으로 나누기 방지)
epsilon = 1e-8

# 먼저 발생률 계산
aha_result['activity_rate'] = aha_result[['Churned', 'Retained']].max(axis=1)

# 최종 Aha Score 계산
aha_result['aha_score'] = (aha_result['gap'] * aha_result['activity_rate']) / (aha_result['p_value'] + epsilon)

# p-value 필터링 (유의미한 것만)
aha_result_final = aha_result[aha_result['p_value'] <= 0.05]

# 최종 우선순위 정렬
aha_result_final = aha_result_final.sort_values('aha_score', ascending=False)

# 결과 확인
aha_result_final[['cluster', 'feature', 'gap', 'activity_rate', 'p_value', 'aha_score']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_aha_dashboard(aha_result_final, top_k=3):
    # 클러스터별 Top3 뽑기
    top3 = (
        aha_result_final
        .sort_values(['cluster', 'aha_score'], ascending=[True, False])
        .groupby('cluster')
        .head(top_k)
        .reset_index(drop=True)
    )

    clusters = top3['cluster'].unique()
    n_clusters = len(clusters)

    # subplot 크기 자동조절
    fig, axes = plt.subplots(1, n_clusters, figsize=(6*n_clusters, 6), sharey=True)
    if n_clusters == 1:
        axes = [axes]

    # 클러스터별 서브플롯
    for i, cluster in enumerate(sorted(clusters)):
        ax = axes[i]

        cluster_data = top3[top3['cluster'] == cluster]

        sns.scatterplot(
            data=cluster_data,
            x="gap",
            y="feature",
            size="aha_score",
            hue="gap",
            palette= sns.color_palette("YlOrBr", as_cmap=True),
            sizes=(200, 1500),
            alpha=0.8,
            ax=ax,
            legend=False
        )

        ax.set_title(f"Cluster {cluster} Aha Moments", fontsize=16)
        ax.set_xlabel("Gap (Retained - Churned)", fontsize=14)
        ax.set_ylabel("Feature", fontsize=14)
        ax.grid(alpha=0.3)

    plt.suptitle("클러스터별 Top3 Aha Moment (오렌지 테마)", fontsize=20)
    plt.tight_layout()
    plt.show()
```
