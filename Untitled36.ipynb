```python
aha_result_final= pd.read_csv('/content/drive/MyDrive/고급프로젝트/aha_result_final.csv')
aha_result_final
```

```python
tops_aha= (aha_result_final.sort_values(['cluster','aha_score'], ascending=[True, False]).groupby('cluster').head(3).reset_index(drop= True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib

def visualize_aha_dashboard(aha_result_final, top_k=3):
    # 클러스터별 Top3 뽑기
    top3 = (
        aha_result_final
        .sort_values(['cluster', 'aha_score'], ascending=[True, False])
        .groupby('cluster')
        .head(top_k)
        .reset_index(drop=True)
    )

    clusters = top3['cluster'].unique()
    n_clusters = len(clusters)

    # subplot 크기 자동조절
    fig, axes = plt.subplots(1, n_clusters, figsize=(6*n_clusters, 6), sharey=True)
    if n_clusters == 1:
        axes = [axes]

    # 클러스터별 서브플롯
    for i, cluster in enumerate(sorted(clusters)):
        ax = axes[i]

        cluster_data = top3[top3['cluster'] == cluster]

        sns.scatterplot(
            data=cluster_data,
            x="gap",
            y="feature",
            size="aha_score",
            hue="gap",
            palette= sns.color_palette("YlOrBr", as_cmap=True),
            sizes=(200, 1500),
            alpha=0.8,
            ax=ax,
            legend=False
        )

        ax.set_title(f"Cluster {cluster} Aha Moments", fontsize=16)
        ax.set_xlabel("Gap (Retained - Churned)", fontsize=14)
        ax.set_ylabel("Feature", fontsize=14)
        ax.grid(alpha=0.3)

    plt.suptitle("클러스터별 Top3 Aha Moment", fontsize=20)
    plt.tight_layout()
    plt.show()
visualize_aha_dashboard(aha_result_final)
```

```python
for idx, row in tops_aha.iterrows():
    print(f"[Cluster {row['cluster']}] {row['feature']}")
    if row['gap'] > 0:
        print(f"유지 유저가 {row['gap']*100:.1f}%p 더 많이 수행")
        print("→ 이 행동을 활성화하기 위한 유도 전략 필요")
    else:
        print(f"유지 유저가 {abs(row['gap'])*100:.1f}%p 덜 수행")
        print("→ 이 행동 감소 또는 문제 해결 필요")
    print()
```
