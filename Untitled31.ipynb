```python
hackle_properties= pd.read_parquet('/content/drive/MyDrive/고급프로젝트/hackle_properties.parquet')
hackle_properties
```

```python
hackle_events= pd.read_parquet('/content/drive/MyDrive/코드잇/hackle_hackle_events.parquet')
hackle_events
```

```python
events = hackle_events[['event_id','event_datetime','event_key','session_id','friend_count','question_id']]
properties = hackle_properties[['id','session_id','user_id','device_id']]

df = events.merge(properties,on='session_id',how='inner')
df.head()
```

```python
# user_id가 ''인 경우는 결측치로 처리
df['user_id'] = df['user_id'].replace('', pd.NA)

# 동일한 session_id + device_id 조합으로 user_id가 있는 값 추출
lookup = (
    df[df['user_id'].notna()]
    .groupby(['session_id', 'device_id'])['user_id']
    .first()
    .reset_index()
)

# 원래 df에 join해서 빈 user_id 채우기
df = df.merge(lookup, on=['session_id', 'device_id'], how='left', suffixes=('', '_fill'))

# user_id가 비어있던 경우에만 fill된 값으로 대체
df['user_id'] = df['user_id'].combine_first(df['user_id_fill'])

# 보조 컬럼 삭제
df.drop(columns=['user_id_fill'], inplace=True)
```

```python

# 이전 값 계산
df['heart_balance_prev'] = df.groupby('session_id')['heart_balance'].shift(1)
df['event_key_prev'] = df.groupby('session_id')['event_key'].shift(1)
df['event_datetime_prev'] = df.groupby('session_id')['event_datetime'].shift(1)

# 변화량 계산
df['heart_balance_diff'] = df['heart_balance'] - df['heart_balance_prev']

# NaN 제거 & 변화 있는 경우 필터링
results_df = df[
    df['heart_balance_diff'].notnull() &
    (df['heart_balance_diff'] != 0)
].copy()

# 컬럼 정리
results_df = result_df[[
    'session_id',
    'event_key_prev', 'event_datetime_prev', 'heart_balance_prev',
    'event_key', 'event_datetime', 'heart_balance',
    'heart_balance_diff'
]]

results_df.head()
```

```python
# 충전 이벤트만 추출 (하트 증가가 있는 경우)
charge_df = results_df[results_df['heart_balance_diff'] > 0].copy()

# 충전 전 이벤트
charge_df['prev_event'] = charge_df['event_key_prev']
# 충전 후 이벤트
charge_df['next_event'] = charge_df['event_key']

# 충전 전후 이벤트 쌍으로 퍼널 구성
funnel_df = charge_df[['session_id', 'event_datetime_prev', 'prev_event', 'event_datetime', 'next_event', 'heart_balance_diff']]
funnel_df
```

```python
# 충전 전 행동 분포
prev_behavior_counts = charge_df['event_key_prev'].value_counts(normalize=True)

# 충전 후 행동 분포
next_behavior_counts = charge_df['event_key'].value_counts(normalize=True)

# 변화량 정리
behavior_change = (next_behavior_counts - prev_behavior_counts).sort_values(ascending=False)
behavior_change
```

```python
# 충전 이벤트만 필터링 (하트 증가 기준)
charge_df = results_df[results_df['heart_balance_diff'] > 0].copy()

# 세션별로 충전 시점 이후 행동 추적
conversion_behaviors = [
    'click_question_open',
    'click_question_start',
    'click_appbar_alarm_center'
]

# 충전 이후 행동 탐지 함수 정의
def detect_post_charge_behavior(df_session):
    df_session = df_session.reset_index(drop=True)
    charge_times = df_session[df_session['heart_balance_diff'] > 0]['event_datetime']
    if charge_times.empty:
        return pd.Series({b: False for b in conversion_behaviors})
    first_charge_time = charge_times.iloc[0]
    post_charge = df_session[df_session['event_datetime'] > first_charge_time]
    return pd.Series({
        b: (post_charge['event_key'] == b).any() for b in conversion_behaviors
    })

# 세션 기준 행동 추적
session_flags = results_df.groupby('session_id').apply(detect_post_charge_behavior).reset_index()

# 세션별 충전 횟수 계산
session_charge_counts = charge_df['session_id'].value_counts().rename('charge_count').reset_index()
session_charge_counts.columns = ['session_id', 'charge_count']
session_flags = session_flags.merge(session_charge_counts, on='session_id', how='left')
session_flags['charge_count'] = session_flags['charge_count'].fillna(0).astype(int)
session_flags['is_recharged'] = session_flags['charge_count'] > 1

# 재충전율 계산
summary = []
for b in conversion_behaviors:
    group = session_flags.groupby(session_flags[b])['is_recharged'].mean()
    summary.append({'behavior': b, 'did_action': True, 'recharge_rate': group.get(True, 0)})
    summary.append({'behavior': b, 'did_action': False, 'recharge_rate': group.get(False, 0)})

recharge_df = pd.DataFrame(summary)
```

```python
# 세션별 시간 순으로 정렬된 event_key 리스트 만들기
session_sequences = (
    hackle_events
    .sort_values(['session_id', 'event_datetime'])
    .groupby('session_id')['event_key']
    .apply(list)
    .reset_index()
)
```

```python
session_sequences['sequence_str']= session_sequences['event_key'].apply(lambda x: '>'.join(x))
```

```python
#문자열 시퀀스를 숫자 벡터로 변환
# TF-IDF란? 는 흔한 행동은 가중치를 줄이고, 드문 행동은 가중치를 높이는 것=> 각 세션을 행동 기반의 고차원 벡터로 바꿔주는 작업
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(session_sequences['sequence_str'])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
session_sequences['cluster'] = kmeans.fit_predict(X)
```

```python
for c in sorted(session_sequences['cluster'].unique()):
    print(f"\nCluster {c} 대표 시퀀스:")
    print(session_sequences[session_sequences['cluster'] == c]['sequence_str'].head(3).to_string(index=False))
```
